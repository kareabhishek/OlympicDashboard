```python
# Checking for missing values
round(df.isna().sum() / df.shape[0] * 100, 2)
```

- *Age*, *Height* and *Weight* are the three columns with missing values less than 30%. Since the missing values in these columns are not high, we will impute the missing values with the median values of the respective columns.
<br>
- The *Medal* column has 85.33% of missing values. These missing values could be because of no medal won by the participant. We will explore this column further and take action accordingly.

```python
# Imputing missing values in Age column with median Age value
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Imputing missing values in Height column with median Height value
df['Height'].fillna(df['Height'].median(), inplace=True)
```

```python
# Imputing missing values in Weight column with median Weight value
df['Weight'].fillna(df['Weight'].median(), inplace=True)
```

```python
# Imputing missing values in Medal column with 'No Medal' value
df['Medal'].fillna('No Medal', inplace=True)
```

```python
# Checking for missing values
df.isna().sum()
```
